```python
import pandas as pd
```

```python
df = pd.read_csv('IndianaUniversityRejects.csv')
df.head()
```

```python
# GRE
totalGre = []
Quants = []
Verbal = []
for i in range(len(df)):
    data = df.iloc[i, 1].split('\n')
    totalGre.append(int(data[1]))
    Quants.append(int(data[3].split(':')[1]))
    Verbal.append(int(data[5].split(':')[1]))
print('We got scores')
```

```python
toeflScore = []
for i in range(len(df)):
    data = df.iloc[i, 2].split('\n')
    val = float(data[1])
    if (val < 10) :
        val = val*10
    toeflScore.append(int(val))
print('Toefl Score fetched')
```

```python
workEx = []
for i in range(len(df)):
    data = df.iloc[i, 3].split('\n')
    if 'NA' in data[1]:
        val = 0
    else:
        val = int(data[1].split(' ')[0])
    workEx.append(val)
print('WorkEx Fetched')
```

```python
Papers = []
for i in range(len(df)):
    data = df.iloc[i, 4].split('\n')
    try:
        if ('National' or 'Local') in data[1]:
            val = 1
        elif 'International' in data[1]:
            val = 2
        else:
            val = 0
    except:
        val = 0
    Papers.append(val)
print('Papers classified')
```

```python
UGScore = []
for i in range(len(df)):
    data = df.iloc[i, 5]
    val = float(data)
    if (val < 10) :
        val = val*10
    UGScore.append(val)
print('UG scores fetched as percentages')
```

```python
IUDict = {'totalGre': totalGre, 'Quants': Quants, 'Verbal': Verbal, 'toeflScore': toeflScore, 'workEx': workEx, 'Papers': Papers, 'UGScore': UGScore}
IUReject = pd.DataFrame(IUDict)
IUReject
```
